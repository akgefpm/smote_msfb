```python
import numpy as np
import pandas as pd
import pickle
import sys
import gc
import os

import snappy
import fastparquet
import matplotlib.pyplot as plt

## Modelling related packages
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn import metrics
from datetime import datetime
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import roc_curve
from matplotlib import pyplot
from sklearn.metrics import precision_recall_curve
from random import choices
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.datasets import make_classification
from scipy.stats import randint as sp_randint
from scipy.stats import uniform as sp_uniform
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
## Summarize the results of all experiments
path = "/domino/datasets/local/CustomerSegmentation/experiment/Imbalance_Ratio_" #0.25/"

exp_data = pd.DataFrame()

for imb_ratio in ['0.20','0.25','0.30','0.35','0.40','0.45']:
    
    folder_path = path + str(imb_ratio)
    
    for ind in range(1,50):
        data_exp1 = pd.read_csv(folder_path+"/Exp_"+str(ind)+"/results.csv")
        data_exp1['imbalance_ratio'] = imb_ratio
        data_exp1['exp_num'] = ind
        exp_data = pd.concat([exp_data, data_exp1], axis=0 )
```

```python
exp_data['f1_score_test_new'] = 1 / ( (1/exp_data['test_precision_new']) + (1/exp_data['test_recall_new']) )
exp_data['f1_score_test_org'] = 1 / ( (1/exp_data['test_precision_org']) + (1/exp_data['test_recall_org']) )
```

```python
exp_data.head(3)
```

```python
# Group by imbalance_ratio and calculate median for both f1 scores
median_df = exp_data.groupby('imbalance_ratio')[['f1_score_test_org', 'f1_score_test_new']].median().reset_index()

# Plot the median F1 scores for both org and new
plt.figure(figsize=(10, 6))
plt.plot(median_df['imbalance_ratio'], median_df['f1_score_test_org'], label='Median Test F1 Score Org', marker='o', linestyle='-', color='b')
plt.plot(median_df['imbalance_ratio'], median_df['f1_score_test_new'], label='Median Test F1 Score New', marker='o', linestyle='-', color='r')

# Set the y-axis limits to be between 0 and 1
plt.ylim(0, 1)

# Add labels and title
plt.title('Median Test F1 Score (Org vs New) Across Varying Imbalance Ratios')
plt.xlabel('Imbalance Ratio')
plt.ylabel('Median Test F1 Score')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Group by imbalance_ratio and calculate median for both recall columns
median_df = exp_data.groupby('imbalance_ratio')[['test_recall_org', 'test_recall_new']].median().reset_index()

# Plot the median values for test_recall_org and test_recall_new
plt.figure(figsize=(10, 6))
plt.plot(median_df['imbalance_ratio'], median_df['test_recall_org'], label='Median Test Recall Org', marker='o', linestyle='-', color='b')
plt.plot(median_df['imbalance_ratio'], median_df['test_recall_new'], label='Median Test Recall New', marker='o', linestyle='-', color='r')

# Set the y-axis limits to be between 0 and 1
plt.ylim(0, 1)

# Add labels and title
plt.title('Median Test Recall (Org vs New) Across Varying Imbalance Ratios')
plt.xlabel('Imbalance Ratio')
plt.ylabel('Median Test Recall')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Group by imbalance_ratio and calculate median for both precision columns
median_df = exp_data.groupby('imbalance_ratio')[['test_precision_org', 'test_precision_new']].median().reset_index()

# Plot the median values for test_precision_org and test_precision_new
plt.figure(figsize=(10, 6))
plt.plot(median_df['imbalance_ratio'], median_df['test_precision_org'], label='Median Test Precision Org', marker='o', linestyle='-', color='b')
plt.plot(median_df['imbalance_ratio'], median_df['test_precision_new'], label='Median Test Precision New', marker='o', linestyle='-', color='r')

# Set the y-axis limits to be between 0 and 1
plt.ylim(0, 1)

# Add labels and title
plt.title('Median Test Precision (Org vs New) Across Varying Imbalance Ratios')
plt.xlabel('Imbalance Ratio')
plt.ylabel('Median Test Precision')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
## Precision Comparison plot

exp_data_melted = exp_data.melt(id_vars=['imbalance_ratio'], value_vars=['test_precision_org', 'test_precision_new'],
                    var_name='Model', value_name='Precision')


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='imbalance_ratio', y='Precision', hue='Model', data=exp_data_melted)

# Set labels and title
plt.title('Model Test Precision - Original Data vs New Sampled Data for varying levels of Imbalance Ratio')
plt.xlabel('Imbalance Ratio')
plt.ylabel('Test Precision')

# Show the plot
plt.show()
```
